# Removing Outliers:

z-Score Technique:

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(cred))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
cred_new=cred[(z<3).all(axis=1)]
print(cred.shape)
print(cred_new.shape)
```

Percentage Data Loss:

```python
percentage_loss=(209593-165752)/209593*100
print(percentage_loss)
```

IQR Technique:

```python
Q1 = cred.quantile(0.25)
Q3 = cred.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
cred_new1= cred[~((cred < (Q1 - 1.5 * IQR)) |(cred > (Q3 + 1.5 * IQR))).any(axis=1)]
print(cred.shape)
print(cred_new1.shape)
```

Percentage Data Loss:

```python
percentage_loss=((209593-61183)/209593)*100
print(percentage_loss)
```

The percentage loss in the z-score is less, so we prefer z-score over IQR.
